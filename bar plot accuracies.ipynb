```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = ['CNN_scratch', 'CNN_ECA', 'CNN_CBAM', 'VGG16', 'VGG19',
          'InceptionV3', 'MobileNetV2', 'ResNet50', 'Xception', 'DenseNet121',
          'ViT-DeiT', 'ViT-Base']
internal_accuracy = [98.4, 80, 87.6, 99, 99, 98.6, 98, 99.2, 97.8, 97.6, 97, 91]
external_accuracy = [86, 80, 88, 94, 96, 95, 93, 94, 94, 93, 91, 76]
```

```python
# Colors
colors = plt.cm.get_cmap('tab20', len(models))

# Bar plot for internal accuracy
plt.figure(figsize=(10, 6))
plt.bar(models, internal_accuracy, color=colors(np.arange(len(models))))
plt.xlabel('Models')
plt.ylabel('Accuracy (%)')
plt.title('Internal Testing Accuracy of Models')
plt.xticks(rotation=45)
plt.ylim(0, 100)
plt.show()
```

```python

# Bar plot for external accuracy
plt.figure(figsize=(10, 6))
plt.bar(models, external_accuracy, color=colors(np.arange(len(models))))
plt.xlabel('Models')
plt.ylabel('Accuracy (%)')
plt.title('External Testing Accuracy of Models')
plt.xticks(rotation=45)
plt.ylim(0, 100)
plt.show()
```
